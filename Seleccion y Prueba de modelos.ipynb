```python
import pandas as pd
import hdbscan
import math
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import silhouette_score
import plotly.express as px


data = pd.read_csv("drive/MyDrive/Tesis/datos_na_v3.csv", encoding='latin1')
data = data.loc[:, ~(data == 'NA').any()]
data = data.dropna()

df = data
df = df[df['zona_geografica'] == 'AMG']
df = df[df['delito'] == 'Robo a persona']

#df = np.array(df[['y','x']])


X = df[['y', 'x']]
X_scaled = np.array(X)


min_cluster_size = 5 # Set the minimum number of samples required to form a cluster
hdbscan_clusterer = hdbscan.HDBSCAN(min_cluster_size=min_cluster_size, metric='haversine')
clusters = hdbscan_clusterer.fit_predict(np.radians(X_scaled))


#plt.scatter(X_scaled[:,0], X_scaled[:,1], c=clusters)

# Create a DataFrame with the coordinates and clusters
df_plot = pd.DataFrame({'x': X_scaled[:, 1], 'y': X_scaled[:, 0], 'cluster': clusters})
#eliminate un-clustered points
#df_plot = df_plot[df_plot['cluster'] != -1]

# Plot the clusters on a map
fig = px.scatter_mapbox(df_plot, lat='y', lon='x', color='cluster',
                        color_discrete_sequence=px.colors.qualitative.D3,
                        zoom=10, height=600)
fig.update_layout(mapbox_style="carto-positron")
fig.show()

silhouette_avg = silhouette_score(np.radians(X_scaled), clusters)
print("The average Silhouette coefficient is :", silhouette_avg)

print(len(np.unique(clusters)))
```
