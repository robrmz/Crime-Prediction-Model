# Data Cleaning y Procesamiento de Datos

```python
#Primero visualizamos la información en nuestra base de datos. 
import pandas as pd
import numpy as np

df=pd.read_csv("datos_full.csv")
df.tail() #Con esto encontramos que hay varias columnas con NA
```

```python
#Para nuestro objetivo, no necesitamos los puntos de datos con NA, por eso los descartamos:
df = df.dropna()
#visualizamos de nueva cuenta
df.tail()
```

Con lo anterior, pasamos de 210705 datos a 210177 de datos 'limpios'

```python
#Visualizamos las diferentes columnas
df.columns
```

```python
#Nos enfocaremos únicamente en los ciertos valores:

#zona_geografica = AMG
df = df[df['zona_geografica'] == 'AMG']
```

```python
#delito = Robo a persona
df = df[df['delito'] == 'Robo a persona']
```

```python
#visualizamos nuestro nuevo set de datos:
df.head()
```

Ahora contamos con 25120 filas de eventos delictivos bajo la categoria de 'Robo a persona'
